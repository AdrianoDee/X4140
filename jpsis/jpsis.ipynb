```python
from ROOT import RooRealVar,RooAbsPdf,RooChebychev,RooExponential,RooGaussian
from ROOT import RooAbsPdf,RooPlot,RooAddPdf,RooDataHist,RooArgSet,RooArgList,RooCBShape
from ROOT import RooVoigtian,RooAbsReal
from ROOT import kGreen,kRed,kBlack,kBlue,kDashed,kDotted,kMagenta
from ROOT.RooFit import Components,LineColor,LineStyle,Name,Normalization,Layout,Format,Label,Parameters,Range,Title
```

```python
massmin = 2.8
massmax = 3.4
jspimean = 3.105
massbins = (massmax - massmin)/histall.GetBinWidth(2)
mass = RooRealVar("mass","JPsi Mass",massmin,massmax)
mass.setBins(int(massbins))
```

```python
histall.SetTitle("JPsi Mass - All; M(#mu#mu)[GeV]; candidates/" + str(histall.GetBinWidth(2)*1000.0) + "MeV")
histall.GetXaxis().SetRangeUser(massmin,massmax)
histall.SetLineColor(kBlue)
histall.SetLineWidth(2)
histall.Draw()
plotname = "plots/jpsiAllplot_" + histallname
c.SaveAs(plotname  + ".png")
c.SaveAs(plotname  + ".eps")
c.SaveAs(plotname  + ".pdf")

histcw.SetTitle("JPsi Mass - CW; M(#mu#mu)[GeV]; candidates/" + str(histcw.GetBinWidth(2)*1000.0) + "MeV")
histcw.GetXaxis().SetRangeUser(massmin,massmax)
histcw.Draw()
plotname = "plots/jpsiCWplot_" + histcwname
c.SaveAs(plotname  + ".png")
c.SaveAs(plotname  + ".eps")
c.SaveAs(plotname  + ".pdf")

histsw.SetTitle("JPsi Mass - SW; M(#mu#mu)[GeV]; candidates/" + str(histsw.GetBinWidth(2)*1000.0) + "MeV")
histsw.GetXaxis().SetRangeUser(massmin,massmax)
histsw.Draw()
plotname = "plots/jpsiSWplot_" + histswname
c.SaveAs(plotname  + ".png")
c.SaveAs(plotname  + ".eps")
c.SaveAs(plotname  + ".pdf")
```

```python
mean = RooRealVar("#mu","mean of jpsi",jspimean,jspimean-0.5,jspimean+0.5)
#sigma = RooRealVar("sigma","width of jpsi",0.0015,0.001,0.02)
sigma1 = RooRealVar("#sigma","width of gaussian1",0.041,0.002,0.065);
sigma2 = RooRealVar("#sigma_{CB}","width of gaussian2",0.02,0.002,0.065);
sigma3 = RooRealVar("#sigma_{3}","width of gaussian3",0.035,0.002,0.055);
a1 = RooRealVar("#alpha_{1}","a1",1.47,-20,20)
n1 = RooRealVar("n1","n1",145.0,125.,150.0)
a2 = RooRealVar("#alpha","a2",1.247,-5.5,5.5)
n2 = RooRealVar("n","n",10.0,0.,165.0)
#signal = RooGaussian("jspsi","jspsi",mass,mean,sigma)
gamma = RooRealVar("gamma","gamma of bw",0.015,0.001,0.05)
#signal = RooVoigtian("signal","signal",mass,mean,gamma,sigma1)
gauss1 = RooGaussian("gauss1","gaussian PDF 1",mass,mean,sigma1)
#gauss2 = RooGaussian("gauss2","gaussian PDF 2",mass,mean,sigma2)
#gauss3 = RooGaussian("gauss3","gaussian PDF 3",mass,mean,sigma3)
#gauss1 = RooCBShape("gauss1","gaussian PDF 1",mass,mean,sigma1, a1, n1);
gauss2 = RooCBShape("gauss2", "gauss2", mass, mean, sigma2, a2, n2);
gaussFrac = RooRealVar("f_{CB}","gaussian fraction",0.65,0.0,1.0)
#gaussFrac2 = RooRealVar("f^{1}_{g}","gaussian fraction2",0.65,0.0,1.0)
signal  = RooAddPdf("model","g1+g2",gauss2,gauss1,gaussFrac)
#signal  = RooAddPdf("model","g1+g2",RooArgList(gauss1,gauss2,gauss3),RooArgList(gaussFrac,gaussFrac2))
```

```python
masslist = RooArgList(mass)
dhall = RooDataHist("dh","dh",masslist,histtofit)
numEvtsAll = dhall.sum(False)
print numEvtsAll
```

```python
signal.fitTo(dhall)
```

```python
cmass = TCanvas("canvas","canvas",1200,1200);
pullpad = TPad("pullpad","pullpad",0.0,0.0,1.0,0.30)
plotpad = TPad("histopad","histopad",0.0,0.35,1.0,1.0)
plotpad.Draw()
pullpad.Draw()
pullpad.SetFillStyle(4004)
plotpad.SetFillStyle(4004)

massFrame = mass.frame()
massFrame.SetTitle("JPsi Mass - CC")
massFrame.GetYaxis().SetTitle("candidates / " + str(histtofit.GetBinWidth(2)*1000.0) + " MeV")
massFrame.GetYaxis().SetTitleOffset(1.28)
massFrame.GetXaxis().SetTitle("M(#mu#mu)[GeV]")
dhall.plotOn(massFrame,Name("Data"))
signal.plotOn(massFrame,Name("Pdf"))
#gauss3.plotOn(massFrame,LineColor(kRed),LineStyle(kDashed),Normalization(gaussFrac2.getValV()*numEvtsAll/(numEvtsAll)))
gauss2.plotOn(massFrame,LineColor(kGreen),LineStyle(kDashed),Normalization(((gaussFrac.getValV())*numEvtsAll),RooAbsReal.NumEvent))#gaussFrac.getValV()*numEvtsAll)/(numEvtsAll)))
gauss1.plotOn(massFrame,LineColor(kMagenta),LineStyle(kDashed),Normalization(((1.0-gaussFrac.getValV())*numEvtsAll),RooAbsReal.NumEvent))
signal.paramOn(massFrame,Layout(0.60,0.99,0.6))
hpull = massFrame.pullHist("Data","Pdf")
#hpull.GetXaxis().SetTitleSize(1.5)
#hpull.GetYaxis().SetTitleSize(1.5)
pullframe = mass.frame(Title("Pull Distribution"))
pullframe.GetXaxis().SetTitleSize(0.04)
pullframe.GetYaxis().SetTitleSize(0.03)
ROOT.gStyle.SetTitleFontSize(0.07)
plotpad.cd()
massFrame.Draw()
pullpad.cd()
pullframe.addPlotable(hpull,"P")
upLine = TLine(massmin,4.0,massmax,4.0)
upLine.SetLineColor(kRed)
upLine.SetLineStyle(kDashed)
upLine.SetLineWidth(2)
dwLine = TLine(massmin,-4.0,massmax,-4.0)
dwLine.SetLineColor(kRed)
dwLine.SetLineStyle(kDashed)
dwLine.SetLineWidth(2)
pullframe.Draw()
dwLine.Draw()
upLine.Draw()
```
